# Matrices

```r
# Generar matriz, a partir de un vector con los datos e indicando la cantidad de filas y columnas
matriz = matrix(data= c(1,2,3,4,5,6,7,8,9,0),
                nrow=2, ncol=5)
matriz
```

```r
matriz[2,3] #Devuelve elemento de la fila 2 y columna 3
matriz[,3] #Devuelte todas las filas de la columna 3
matriz[1:2,3] #Devuelve de la fila 1 hasta 2 y columna 3
# matriz(1:2,1) #Devuelve de la fila hasta 2 y todas las columnas
```

Tipos de variables

1. Variable numerica
2. Variable texto
3. Variable categorica

```r
# Definicir un vector de variable categorica
# Se deben definir las categorias y los valores.

Vector3 = factor(levels = c("a","b","c","d"), x = c("b","b","c","c","b","c","a","c","b","c","b","b","b"))
Vector3
```

```r
# Devuelve la frecuencia absoluta basada en las categorias de la variable
conteo = table(Vector3)
conteo
```

```r
# Devuelve la frecuencia relativa de la tabla conteo de un vector categorico
prop.table(conteo)
```

```r
# si al definir la variable categorica con factor() no se le definen los levels
# la frecuencia absoluta y relativa se contruye con los valores existentes.
# si en los datos hay un valor que no esta definido en los levels
# R lo trabaja como un NA
```

## Leer archivo csv de la web

```r
url = "https://raw.githubusercontent.com/chris0162/datasets/main/Dataset%20Diabetes/diabetes_V2.csv"
diabetes = read.csv(url, row.name=1)
head(diabetes)
```

## Seleccionar columnas numericas usando SELECT + WHERE + HEAD

```r
# Seleccionar columnas numericas usando SELECT + WHERE + HEAD
diabetes %>%
  select(where(is.numeric)) %>%
  head
```

## Seleccionar solo columnas numericas

```r
# Seleccionar solo columnas numericas
a = diabetes %>% select(where(is.numeric))
a %>% head
```

## Seleccionar columnas NO numericas

```r
# Seleccionar columnas NO numericas
diabetes %>%
  select(-where(is.numeric)) %>%
  head()
```

## Ocular una columna de la tabla usando SELECT


```r
# Ocular una columna de la tabla usando SELECT
diabetes %>%
  select(-colesterol) %>%
  head
```

## Uso de operador pipe %>%

```r
# Uso de operador pipe %>%
  scale(head(select(diabetes, altura, peso)))
```

```r
# Uso de operador pipe %>%

diabetes %>%
  select(altura,peso) %>%
  scale() %>%
  head()
```

## Seleccionar columnas cuyo nombre inicie con a

```r
# Seleccionar columnas cuyo nombre inicie con a
diabetes %>%
  select(starts_with("a")) %>%
  head()
```

## Seleccionar columnas cuyo nombre termine con a

```r
# Seleccionar columnas cuyo nombre termine con a
diabetes %>%
  select(ends_with("a")) %>%
  head()
```

## Filtrar filas que cumplen con un criterio

```r
# Filtrar filas que cumplen con un criterio
diabetes %>%
  filter(altura > 150)
```

## Calcula la frecuencia absoluta con COUNT()

```r
# Calcula la frecuencia absoluta con COUNT()
diabetes %>% count(genero)
```

## Eliminar datos nulos

```r
# Eliminar datos nulos
diabetes2 = na.omit(diabetes)
```

## Calcula la media de las variables numericas

```r
# Calcula la media de las variables numericas
diabetes2 %>%
  summarise(across(where(is.numeric),mean))
```
